```python
# Importar el módulo JSON. (Hecho por Nicole y Evangelina)
import json

import random as rd
class Persona:
  def __init__(self, nombre="", apellido="", edad=0, dni=0): # constructor con parametros
    self.__nombre = nombre
    self.__apellido = apellido
    self.__edad = edad
    self.__dni = dni

  @property
  def nombre(self):
    return self.__nombre
  @property
  def apellido(self):
    return self.__apellido
  @property
  def edad(self):
    return self.__edad
  @property
  def dni(self):
    return self.__dni

  @nombre.setter
  def nombre(self, nombre):
    self.__nombre = nombre
  @apellido.setter
  def apellido(self, apellido):
    self.__apellido = apellido
  @edad.setter
  def edad(self, edad):
    self.__edad = edad
  @dni.setter
  def dni(self, dni):
    self.__dni = dni

  def __str__(self):
    return f"Nombre: {self.nombre}, Apellido: {self.apellido}, edad: {self.edad}, dni: {self.dni}"

#pedro = Persona("Pedro", "Martinez",24, 37654321)
#maria = Persona("Maria","Liendro", 20, 40123456)
#print(pedro)
#print(maria)

class Maestro(Persona):
  def __init__(self, nombre, apellido, edad, dni, legajo, materias):
    super().__init__(nombre, apellido, edad, dni)
    self.__legajo = legajo
    self.__materias = materias

  def get_legajo(self):
    return self.__legajo
  def get_materias(self):
    return self.__materias

  def set_legajo(self, legajo):
    self.__legajo = legajo
  def set_materia(self, materia):
    self.__materias.append(materia)
    print(f"La materia {materia} se agrego con éxito.")
  def eliminar_materia(self, materia):
    if materia in self.__materias:
      self.__materias.remove(materia)
      print(f"La materia {materia} fue eliminada")
    else:
      print(f"El docente no dicta la materia {materia}")
  def __str__(self):
    return super().__str__() + f", legajo: {self.get_legajo()}, Materias: {self.get_materias()}"

#maestro1 = Maestro("Luis", "Garcia", 49, 25321321, 9876, ["Lengua", "matemática", "Historia", "Geografía"])
#print(maestro1)
#maestro1.set_materia("Ingles")
#print(maestro1)
# tarea hacer un metodo que elimine una materia de la clase Maestro

class Alumno(Persona):
  def __init__(self,  nombre, apellido, edad, dni, grado, turno, materias):
    super().__init__(nombre, apellido, edad, dni)
    self.__grado = grado
    self.__turno = turno
    self.__materias = materias
    self.__notas = []
    self.__comision = "No asignada"

  def get_grado(self):
    return self.__grado
  def get_turno(self):
    return self.__turno
  def get_comision(self):
    return self.__comision
  def generar_notas(self, cant):
    for i in range(cant):
      self.__notas.append(rd.randint(0,10))

  def mostrar_notas(self):
    if len(self.__notas) > 0:
      cadena = ""
      for i in range(len(self.__materias)):
        cadena = cadena + self.__materias[i] + ": " + str(self.__notas[i]) + "\n"
      return cadena
    else:
      return "No hay notas cargadas"
  def promedio(self):
    if len(self.__notas) > 0:
      return sum(self.__notas) / len(self.__notas)
    else:
      print("No hay notas cargadas")
      return 0
  def asignar_comision(self, prom):
    if self.edad < 17:
      if prom >= 8:
        self.__comision = "A"
      elif prom >= 6:
        self.__comision = "B"
      else:
        self.__comision = "C"
    else:
      self.__comision = "N"
  def set_grado(self, grado):
    self.__grado = grado
  def set_turno(self, turno):
    self.__turno = turno

  def __str__(self):
    return "Alumno/a: " + super().__str__() + f", Grado {self.get_grado()}, turno: {self.get_turno()}, Comision: {self.get_comision()}\nNotas: {self.mostrar_notas()}\n"

  # Crear un diccionario en donde la clave sea los atributos de la clase Alumno, y los valores serán los datos guardados en dichos atributos. (Hecho por Nicole)
  def crear_diccionario(self):
    diccionario = {
      "nombre": self.nombre,
      "apellido": self.apellido,
      "edad": self.edad,
      "dni": self.dni,
      "grado": self.get_grado(),
      "turno": self.get_turno(),
      "materias": self.__materias,
      "notas": self.__notas,
      "comision": self.get_comision()
    }
    return diccionario
class Preceptor(Persona):
  def __init__(self, nombre, apellido, edad, dni):
    super().__init__(nombre, apellido, edad, dni)
    self.__comision = []
    self.__lista_alumnos = []

  def get_comision(self):
    return self.__comision
  def get_lista_alumnos(self):
    return self.__lista_alumnos

  def set_comision(self, comision):
    self.__comision.append(comision)
  def eliminar_comision(self, comision):
    if comision in self.__comision:
      self.__comision.remove(comision)
      print(f"La comision {comision} fue eliminada")
    else:
      print(f"El/la preceptora no está asignada a la comision {comision}")
  def set_agregar_alumno(self, alumno):
    if not alumno in self.__lista_alumnos:
      self.__lista_alumnos.append(alumno)
    else:
      print("el alumno ya está en la lista")
  def set_eliminar_alumno(self, alumno):
    if alumno in self.__lista_alumnos:
      self.__lista_alumnos.remove(alumno)
      print("Alumno eliminado con éxito.")
    else:
      print(f"El alumno que desea eliminar no está  en la lista")

  def mostrar_lista_alumnos(self):
    if len(self.__lista_alumnos) > 0:
      for i in range(len(self.__lista_alumnos)):
        print(self.__lista_alumnos[i])
    else:
      print("La lista de alumnos está vacía.")
  def __str__(self):
    return "Preceptor/a: " + super().__str__() + f", Comisiones asignadas {self.get_comision()}"



#preceptor1 = Preceptor("Maria", "Carreras", 29, 33888444)
#preceptor1.set_comision("A")
#preceptor1.set_comision("B")
#print(preceptor1)
#alumno2 = Alumno("Lisa", "Simpson", 9, 44321543, 4, "mañana", ["Lengua", "matemática", "Historia", "Geografía"])
#alumno2.asignar_comision(alumno2.promedio())
#preceptor1.set_agregar_alumno(alumno1)
#preceptor1.set_agregar_alumno(alumno2)
#preceptor1.mostrar_lista_alumnos()
#preceptor1.eliminar_comision("A")
#print(preceptor1)


# Definir una función para convertir el archivo JSON a diccionario. (Hecho por Nicole)
def json_a_diccionario(nombre_archivo):
    with open(nombre_archivo, 'r') as f:
        contenido = f.read()
        diccionario = json.loads(contenido)
    print("Conversion de JSON a diccionario realizado con éxito!")
    return diccionario

# Definir una función para convertir el diccionario en un archivo JSON. (Hecho por Evangelina)
def diccionario_a_json(diccionario, nombre_archivo):
    with open(nombre_archivo, 'w') as f:
        json.dump(diccionario, f, indent=4, ensure_ascii=False)
    print("Conversion de diccionario a JSON realizado con éxito!")

# Mostrar el diccionario y el archivo JSON. (Hecho por Evangelina)
def mostrar_resultados(diccionario, archivo_json):
    print("Diccionario:")
    print(type(diccionario))
    print(diccionario)
    print("Archivo JSON:")
    contenido = open(archivo_json).read()
    print(type(contenido))
    print(contenido)


alumno = Alumno("Bartolomeo", "Simpson", 18, 47321456, 5, "mañana", ["Lengua", "matemática", "Historia", "Geografía"])

alumno.generar_notas(4)
alumno.asignar_comision(alumno.promedio())

archivo_json = "alumno.json"

print("\nJSON A DICCIONARIO")
diccionario = json_a_diccionario(archivo_json)

print("\nJSON A DICCIONARIO")
diccionario = json_a_diccionario(archivo_json)

print("\nJSON A DICCIONARIO")
mostrar_resultados(diccionario, archivo_json)
```
